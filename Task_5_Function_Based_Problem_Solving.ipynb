```python
def analyze_post_engagement(daily_likes):
    total_likes = 0
    # Use loop for calculations
    for likes in daily_likes:
        total_likes += likes

    # Use conditional logic
    if total_likes >= 1000:
        post_status = "Viral Post"
    else:
        post_status = "Normal Engagement"

    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {post_status}")

# Example
likes_data = [250, 300, 500]  # Total is 1050
analyze_post_engagement(likes_data)

```

```python
def medicine_stock_alert(stock_level):

    # Use conditional logic
    if stock_level < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print(f"Medicine Stock: {stock_level}")
    print(f"Status: {status}")

# Example Usage
current_stock = 6
medicine_stock_alert(current_stock)

```

```python
def rainfall_adequacy_checker(seasonal_readings, required_level):

    total_rainfall = 0

    # Use loop and math operations
    for rainfall in seasonal_readings:
        total_rainfall += rainfall

    if not seasonal_readings:
        average_rainfall = 0
    else:
        average_rainfall = total_rainfall / len(seasonal_readings)

    # Conditional logic
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    # Formatting the output to match the expected format
    print(f"Average Rainfall: {int(average_rainfall)}")
    print(f"Rainfall Status: {status}")

# Example Usage
readings = [80, 65, 71, 72] # Average is 72
required_minimum = 70
rainfall_adequacy_checker(readings, required_minimum)
```

```python
def duplicate_account_detection(usernames):

    # Use set data structure
    unique_usernames = set(usernames)

    # Conditional logic to check for duplicates
    if len(usernames) != len(unique_usernames):
        duplicates_found = "Yes"
    else:
        duplicates_found = "No"

    print(f"Duplicate Accounts Found: {duplicates_found}")

# Example Usage
account_list = ["user123", "gemini_ai", "user123", "alpha_bet"]
duplicate_account_detection(account_list)
```

```python
def appointment_eligibility_checker(patient_age):

    # Use condition checks
    if patient_age >= 18:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print(f"Patient Age: {patient_age}")
    print(f"Eligibility Status: {eligibility}")

# Example Usage
age = 21
appointment_eligibility_checker(age)
```

```python
def premium_crop_price_filter(crop_prices):

    premium_crops = []
    threshold = 2000

    # Use list filtering logic (as hinted)
    for price in crop_prices:
        if price > threshold:
            premium_crops.append(price)

    print(f"Premium Crops: {premium_crops}")

# Example Usage
prices = [1500, 2500, 1900, 3200, 800]
premium_crop_price_filter(prices)
```

```python
def application_health_checker(error_count):

    # Use conditional statements (as hinted)
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5: # This covers 1 to 5
        status = "Minor Issues"
    else: # This covers > 5
        status = "Critical Issues"

    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")

# Example Usage
errors = 7
application_health_checker(errors)

```

```python
DAILY_LIMIT = 50000

def daily_transaction_limit_checker(transaction_amount):

    # Use conditional logic (as hinted)
    if transaction_amount <= DAILY_LIMIT:
        status = "Approved"
    else:
        status = "Rejected"

    print(f"Transaction Amount: {transaction_amount}")
    print(f"Transaction Status: {status}")

# Example Usage
amount = 60000
daily_transaction_limit_checker(amount)

```

```python
def student_attendance_eligibility(attended_classes, total_classes):

    # Use list and loop concept (implied in the calculation)
    if total_classes == 0:
        attendance_percentage = 0.0
    else:
        # Math operation to calculate percentage
        attendance_percentage = (attended_classes / total_classes) * 100

    # Conditional logic
    if attendance_percentage >= 75.0:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    # Formatting output to one decimal place for the percentage
    print(f"Attendance Percentage: {attendance_percentage:.1f}")
    print(f"Exam Eligibility: {eligibility}")

# Example Usage
attended = 80
total = 100
student_attendance_eligibility(attended, total)
```

```python
def smart_electricity_bill_analyzer(units_consumed):

    bill = 0
    remaining_units = units_consumed

    # Billing Calculation - Use conditional logic for slabs (as hinted)

    # Slab 1: First 100 units @ ₹3 per unit
    if remaining_units > 100:
        bill += 100 * 3
        remaining_units -= 100
    else:
        bill += remaining_units * 3
        remaining_units = 0

    # Slab 2: Next 100 units (101-200) @ ₹5 per unit
    if remaining_units > 100:
        bill += 100 * 5
        remaining_units -= 100
    elif remaining_units > 0:
        bill += remaining_units * 5
        remaining_units = 0

    # Slab 3: Above 200 units @ ₹7 per unit
    if remaining_units > 0:
        bill += remaining_units * 7

    # Usage Classification - Use conditional logic for classification
    if bill < 500:
        usage_status = "Low Usage"
    elif 500 <= bill <= 1500:
        usage_status = "Moderate Usage"
    else: # bill > 1500
        usage_status = "High Usage"

    # Use return values (as hinted)
    return bill, usage_status

# Example Usage (250 units: 100*3 + 100*5 + 50*7 = 300 + 500 + 350 = 1150)
units = 250
total_bill, usage = smart_electricity_bill_analyzer(units)

print(f"Units Consumed: {units}")
print(f"Total Bill: ₹{total_bill}")
print(f"Usage Classification: {usage}")
```
